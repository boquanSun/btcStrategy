```python
import pandas as pd
```

```python
basis_200626 = pd.read_csv('.\edav_onchaindata\ 200626 .csv')
basis_200626["date"] = pd.to_datetime(basis_200626["date"])
basis_200626.head()
```

```python
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(specs=[[{"secondary_y": True}]])


fig.add_trace(
    go.Scatter(x=basis_200626["date"], y=basis_200626["annual_basis"]),
   secondary_y=False
)


fig.add_trace(
    go.Scatter(x=basis_200626["date"], y=basis_200626["fear"]),
    secondary_y=True
)


fig.show()
```
